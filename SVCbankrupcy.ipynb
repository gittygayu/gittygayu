```python
import pandas as pd
bankrp=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Bankruptcy.csv",sep=";")
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
from sklearn.svm import SVC

dum_bankrp = pd.get_dummies(bankrp,drop_first=True)
X = dum_bankrp.iloc[:,0:-1]
y = dum_bankrp.iloc[:,-1]


X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, 
                                                    random_state=2021,
                                                    stratify=y)
#############linearr##############
svc = SVC(probability = True,kernel='linear',C=2)
fitSVC = svc.fit(X_train, y_train)
y_pred = fitSVC.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

print(accuracy_score(y_test, y_pred))
```
